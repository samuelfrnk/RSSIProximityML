```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Experiment_3.csv')
average_rssi_per_distance = df.groupby('Distance')['RSSI-Value'].mean().reset_index()

def rssi_from_distance(distance, A, beta):
    return A - 10 * beta * np.log10(distance)

A = -64.5
beta = 1.8

distances = np.linspace(0.01, 10, 100)

rssi_model = rssi_from_distance(distances, A, beta)

plt.figure(figsize=(10, 6))
plt.plot(average_rssi_per_distance['Distance'], average_rssi_per_distance['RSSI-Value'], label='Measured RSSI', marker='o', linestyle='-')
plt.plot(distances, rssi_model, label='Path-Loss Model', linestyle='--')

plt.xlabel('Distance (m)')
plt.ylabel('RSSI Value (dBm)')
plt.title('Average RSSI Value for Each Distance with Path-Loss Model')
plt.legend()
plt.xlim(2, 10)
plt.grid(True)
plt.show()

```
